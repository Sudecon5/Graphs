```python
import seaborn as sns

# Given DataFrame
df = pd.DataFrame({
    "Company": ["APPL", "MSFT", "RIL", "TISCO", "Mahindra & Mahindra", "TESLA"],
    "Stock Price": [155.7, 105.3, 81.89, 90.99, 56.65, 78.89]
})

# Plot line graph with Seaborn
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x="Company", y="Stock Price", marker='o', color='purple', linewidth=2.5, markersize=10)

# Customize the plot
plt.title("Stock Prices of Companies", fontsize=14)
plt.xlabel("Company", fontsize=12)
plt.ylabel("Stock Price (USD)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# DataFrame with Electronics and Groceries sales
df_1 = pd.DataFrame({
    "Category": ["Electronics"] * 6 + ["Groceries"] * 5,  # Define categories
    "Product": ["TV", "Refrigerator", "Air Conditioner", "TV", "Refrigerator", "Air Conditioner", 
                "Milk", "Bread", "Eggs", "Fruits", "Vegetables"],
    "Sales": [225, 389, 300, 300, 119, 411,  # Electronics sales
              120, 78, 67, 100, 50]          # Groceries sales (example values)
})
print(df_1.to_string(index=False))
```

```python
grouped = df_1.groupby(["Category", "Product"])["Sales"].sum().unstack()
print(grouped)
```

```python
grouped.plot(kind="bar", stacked=True, figsize=(7, 6))

# Customize
plt.title("Sales Composition by Category", fontsize=14)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.xticks(rotation=0)  # Keep category labels horizontal
plt.legend(title="Product", bbox_to_anchor=(1, 1)) 
plt.grid(axis="y", linestyle="--", alpha=0.2)

# Show
plt.tight_layout()
plt.show()
```

```python
GDP = pd.DataFrame({
    "Agriculture":[256,467,500,345,525,456],
    "Manufaturing":[788,671,571,678,728,678],
    "IT":[872,716,571,681,561,456],
    "Miscellenous":[1118,819,919,1001,1990,891]})
GDP
```

```python
sector_totals = GDP.sum()
print(sector_totals)
```

```python
plt.figure(figsize=(5, 5))
#colors = ["#4CAF50", "#2196F3", "#FF9800", "#9C27B0"]  # Green, Blue, Orange, Purple
explode = (0.01, 0.01, 0.05, 0.01)  # Slightly separate slices

# Create pie chart
plt.pie(
    sector_totals,
    labels=sector_totals.index,
    autopct="%1.1f%%", # Show percentages
    startangle=140,  # Rotate pie for better readability
    shadow=True,        # Add shadow for depth
    textprops={"fontsize": 12},
    explode= explode
)

# Add title
plt.title("GDP Contribution by Sector (Total Across 1st Quarter)", fontsize=16, pad=20)

# Show plot
plt.tight_layout()
plt.show()
```
